Мы должны вернуть только тех пользователей, чей возраст выше нижнего параметра low и ниже верхнего параметра high. Распечатайте их адрес электронной почты и возраст.
```python
try:
    with open('users.txt','r') as file:
        lines = file.readlines()
except FileNotFoundError:
    print('the file that you would like to read is not found')
except Exception:
    print('some error occured')
```

```python
class Users:
    def __init__(self, имя, фамилия, почта, возраст):
        self.имя = имя
        self.фамилия = фамилия
        self.почта = почта
        self.возраст = возраст
    def getAge(self):
        return 2023 - self.возраст
```

```python
user_list = []
for line in lines:
    строки = line.strip().split(',')
        if len(строки) >=5:
            имя, фамилия, почта, возраст = строки[1:]
            user = Users(имя, фамилия, почта, int(возраст))
            user_list.append(user)
```

```python
def filter(users,low= 30, high = 50):
    for user in users:
        возраст = user.getAge()
        if low <= возраст <= high:
            print(f"{user.почта}, {возраст}")
```
